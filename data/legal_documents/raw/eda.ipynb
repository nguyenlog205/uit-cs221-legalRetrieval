```python
import pandas as pd
import glob
import os
```

```python
folder_path = '.'  
# Hoặc thử đường dẫn tương đối như trong ảnh bạn đang thử:
# folder_path = os.path.join("..", "data", "legal_documents", "raw")

# 2. Lấy danh sách tất cả file .json
json_files = glob.glob(os.path.join(folder_path, "*.json"))

# 3. Đọc và gộp dữ liệu
all_data = []
for file in json_files:
    try:
        # Đọc từng file json thành DataFrame
        df = pd.read_json(file)
        all_data.append(df)
    except ValueError as e:
        print(f"Lỗi đọc file {file}: {e}")

# Gộp tất cả thành 1 bảng lớn
if all_data:
    final_df = pd.concat(all_data, ignore_index=True)
    
    # 4. Lấy cột link (Thay 'Link chi tiết' bằng tên key thực tế trong file json của bạn)
    # Kiểm tra xem cột nào chứa link bằng lệnh: print(final_df.columns)
    if 'Link chi tiết' in final_df.columns:
        list_links = final_df['Link chi tiết'].tolist()
        print(f"Tìm thấy {len(list_links)} links.")
        print(list_links[:5]) # In thử 5 link đầu
    else:
        print("Không tìm thấy cột 'Link chi tiết'. Hãy kiểm tra lại tên cột.")
```
